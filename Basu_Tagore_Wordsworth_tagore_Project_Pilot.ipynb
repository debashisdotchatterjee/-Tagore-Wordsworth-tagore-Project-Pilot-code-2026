```python
# ============================================================
# Synthetic Demonstration for:
# Human--Algorithm Affective Drift through Tagore and Wordsworth
# Statistical + AI/ML Distance-Based Methodology
# ============================================================
# What this code demonstrates:
# 1. Synthetic human and AI response-vector data generation
# 2. Tagore vs Wordsworth exposure across 0,3,6,9,12,15 checkpoints
# 3. AI vs human and proprietary vs open/API model comparison
# 4. ML/NLP-style feature extraction represented synthetically
# 5. Emotional Representative Distance (ERD)
# 6. Affective Drift Index (ADI)
# 7. Mixed-effects modelling / robust OLS fallback
# 8. PERMANOVA-style multivariate group separation test
# 9. Benjamini-Hochberg FDR correction
# 10. Saved tables, figures, README, and ZIP output
# ============================================================

# ----------------------------
# 0. Install / import packages
# ----------------------------

import os, sys, zipfile, warnings, textwrap, subprocess, datetime
warnings.filterwarnings("ignore")

def ensure_package(pkg, import_name=None):
    import_name = import_name or pkg
    try:
        __import__(import_name)
    except ImportError:
        print(f"Installing {pkg} ...")
        subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", pkg])

ensure_package("numpy")
ensure_package("pandas")
ensure_package("matplotlib")
ensure_package("scipy")
ensure_package("scikit-learn", "sklearn")
ensure_package("statsmodels")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.spatial.distance import pdist, squareform
from scipy.stats import ttest_ind
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from statsmodels.stats.multitest import multipletests
import statsmodels.formula.api as smf

try:
    from IPython.display import display, FileLink
except Exception:
    display = print

# ----------------------------
# 1. Output folders
# ----------------------------

RUN_TAG = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
BASE_DIR = f"/content/affective_ai_synthetic_demo_{RUN_TAG}"
TABLE_DIR = os.path.join(BASE_DIR, "tables")
FIG_DIR = os.path.join(BASE_DIR, "figures")
MODEL_DIR = os.path.join(BASE_DIR, "model_outputs")

for d in [BASE_DIR, TABLE_DIR, FIG_DIR, MODEL_DIR]:
    os.makedirs(d, exist_ok=True)

print("Output folder:", BASE_DIR)

# ----------------------------
# 2. Global simulation settings
# ----------------------------

np.random.seed(2026)

CHECKPOINTS = np.array([0, 3, 6, 9, 12, 15])
TRADITIONS = ["Tagore", "Wordsworth"]
AXES = ["Transcendental", "Relational", "Environmental"]
SOURCE_MODELS = ["Human", "Claude_Proprietary", "Open_Model"]

N_HUMAN_PER_CELL = 10     # 2 traditions x 3 axes x 10 = 60 human participants
N_AI_RUNS_PER_CELL = 8    # repeated AI runs per model/tradition/axis
N_SCENARIOS = 4

SCENARIOS = [
    "S0_Control_Minimal_Drift",
    "S1_Human_Tagore_Cultural_Drift_AI_Flattening",
    "S2_Claude_Snapback_Open_Sustained_Drift",
    "S3_Open_Model_Unstable_Semantic_Drift"
]

# ----------------------------
# 3. Feature blocks
# ----------------------------
# The features mimic real outputs from NLP/ML pipelines:
# lexical, syntactic, VAD/emotion, embedding, and cultural-affective coding.

FEATURE_BLOCKS = {
    "Lexical": [
        "lexical_density",
        "moving_average_type_token_ratio",
        "vocabulary_richness"
    ],
    "Syntactic": [
        "mean_sentence_length",
        "punctuation_rate",
        "clause_density"
    ],
    "Affective_VAD": [
        "valence",
        "arousal",
        "dominance"
    ],
    "Fine_Emotion": [
        "emotion_joy",
        "emotion_sadness",
        "emotion_fear",
        "emotion_anger",
        "emotion_awe",
        "emotion_longing"
    ],
    "Embedding": [f"embedding_{j:02d}" for j in range(1, 13)],
    "Rasa_Bhava": [
        "rasa_sringara",
        "rasa_karuna",
        "rasa_shanta",
        "rasa_bhakti",
        "viraha_longing",
        "monsoon_affect"
    ]
}

FEATURES = []
for block_features in FEATURE_BLOCKS.values():
    FEATURES.extend(block_features)

FEATURE_TO_BLOCK = {}
for block, feats in FEATURE_BLOCKS.items():
    for f in feats:
        FEATURE_TO_BLOCK[f] = block

print(f"Total synthetic features: {len(FEATURES)}")
print("Feature blocks:", {k: len(v) for k, v in FEATURE_BLOCKS.items()})

# ----------------------------
# 4. Helper functions
# ----------------------------

def make_zero_vector():
    return pd.Series(0.0, index=FEATURES)

def normalize_vector(v):
    v = pd.Series(v, index=FEATURES).astype(float)
    norm = np.linalg.norm(v.values)
    if norm == 0:
        return v
    return v / norm

def axis_base_shift(axis):
    """
    Baseline tendencies by affective axis.
    These create mild differences in initial response style.
    """
    v = make_zero_vector()
    if axis == "Transcendental":
        v["emotion_awe"] += 0.6
        v["rasa_shanta"] += 0.5
        v["rasa_bhakti"] += 0.4
        v["dominance"] -= 0.2
        v["embedding_01"] += 0.4
        v["embedding_02"] += 0.3
    elif axis == "Relational":
        v["emotion_sadness"] += 0.4
        v["emotion_longing"] += 0.6
        v["viraha_longing"] += 0.5
        v["valence"] -= 0.3
        v["embedding_03"] += 0.4
        v["embedding_04"] += 0.3
    elif axis == "Environmental":
        v["emotion_awe"] += 0.3
        v["monsoon_affect"] += 0.6
        v["arousal"] += 0.2
        v["embedding_05"] += 0.5
        v["embedding_06"] += 0.3
    return v

def tradition_direction(tradition, axis):
    """
    Main drift direction caused by repeated exposure.
    This is synthetic, but conceptually tied to the proposal:
    Tagore receives stronger Rasa/Bhava-oriented cultural-affective movement,
    while Wordsworth receives stronger Romantic nature/reflection cues.
    """
    v = make_zero_vector()

    # Common repeated-exposure effect
    v["moving_average_type_token_ratio"] += 0.25
    v["mean_sentence_length"] += 0.20
    v["clause_density"] += 0.15
    v["emotion_awe"] += 0.15

    if tradition == "Tagore":
        v["rasa_bhakti"] += 0.65
        v["rasa_shanta"] += 0.50
        v["viraha_longing"] += 0.45
        v["rasa_karuna"] += 0.35
        v["emotion_longing"] += 0.35
        v["embedding_07"] += 0.55
        v["embedding_08"] += 0.45
        if axis == "Environmental":
            v["monsoon_affect"] += 0.70
            v["embedding_09"] += 0.35
        if axis == "Relational":
            v["rasa_sringara"] += 0.35
            v["viraha_longing"] += 0.45
        if axis == "Transcendental":
            v["rasa_bhakti"] += 0.40
            v["rasa_shanta"] += 0.35

    if tradition == "Wordsworth":
        v["emotion_awe"] += 0.45
        v["valence"] += 0.20
        v["dominance"] += 0.10
        v["embedding_01"] += 0.45
        v["embedding_05"] += 0.40
        v["embedding_10"] += 0.35
        if axis == "Environmental":
            v["emotion_joy"] += 0.35
            v["embedding_11"] += 0.45
        if axis == "Relational":
            v["emotion_sadness"] += 0.35
            v["emotion_longing"] += 0.25
        if axis == "Transcendental":
            v["emotion_awe"] += 0.35
            v["rasa_shanta"] += 0.15

    return normalize_vector(v)

def source_modifier(source_model, tradition, scenario, checkpoint):
    """
    Returns drift strength multiplier and noise scale.
    """
    x = checkpoint / 15.0

    # Base drift shape: cumulative exposure
    drift_curve = x

    # Default strength and noise
    strength = 0.85
    noise = 0.20

    if source_model == "Human":
        strength = 0.95
        noise = 0.28

    if source_model == "Claude_Proprietary":
        strength = 0.80
        noise = 0.18

    if source_model == "Open_Model":
        strength = 0.90
        noise = 0.25

    if scenario == "S0_Control_Minimal_Drift":
        strength = 0.10
        noise = 0.18

    elif scenario == "S1_Human_Tagore_Cultural_Drift_AI_Flattening":
        if source_model == "Human" and tradition == "Tagore":
            strength = 1.35
            noise = 0.26
        elif source_model != "Human" and tradition == "Tagore":
            strength = 0.55
            noise = 0.18
        elif source_model == "Human" and tradition == "Wordsworth":
            strength = 0.95
            noise = 0.25
        else:
            strength = 0.75
            noise = 0.18

    elif scenario == "S2_Claude_Snapback_Open_Sustained_Drift":
        if source_model == "Claude_Proprietary":
            # early drift, then partial snapback after checkpoint 9
            if checkpoint <= 9:
                drift_curve = checkpoint / 9.0
            else:
                drift_curve = max(0.45, 1.0 - 0.55 * ((checkpoint - 9) / 6.0))
            strength = 1.00
            noise = 0.16
        elif source_model == "Open_Model":
            drift_curve = x
            strength = 1.10
            noise = 0.25
        else:
            strength = 0.95
            noise = 0.25

    elif scenario == "S3_Open_Model_Unstable_Semantic_Drift":
        if source_model == "Open_Model":
            strength = 1.15
            noise = 0.45
        elif source_model == "Claude_Proprietary":
            strength = 0.75
            noise = 0.15
        else:
            strength = 0.95
            noise = 0.25

    return strength * drift_curve, noise

def random_semantic_instability(source_model, scenario, checkpoint):
    """
    Extra random embedding movement, especially for unstable open-model scenario.
    """
    v = make_zero_vector()
    if scenario == "S3_Open_Model_Unstable_Semantic_Drift" and source_model == "Open_Model" and checkpoint > 0:
        for f in FEATURE_BLOCKS["Embedding"]:
            v[f] += np.random.normal(0, 0.40 * checkpoint / 15.0)
    return v

def simulate_one_unit(scenario, source_model, tradition, axis, unit_id):
    """
    Simulate repeated responses for one human participant or one AI model-run.
    """
    base = make_zero_vector()

    # Baseline axis style
    base += axis_base_shift(axis)

    # Mild source-specific baseline
    if source_model == "Human":
        base["moving_average_type_token_ratio"] += 0.10
        base["emotion_longing"] += 0.05
    elif source_model == "Claude_Proprietary":
        base["lexical_density"] += 0.08
        base["mean_sentence_length"] += 0.10
    elif source_model == "Open_Model":
        base["punctuation_rate"] += 0.05
        base["arousal"] += 0.05

    # Unit-specific random baseline variation
    base = base + pd.Series(np.random.normal(0, 0.35, len(FEATURES)), index=FEATURES)

    drift_dir = tradition_direction(tradition, axis)

    rows = []
    for ckpt in CHECKPOINTS:
        mult, noise_sd = source_modifier(source_model, tradition, scenario, ckpt)

        # Synthetic feature vector
        z = base + mult * 2.8 * drift_dir

        # Scenario-specific semantic instability
        z += random_semantic_instability(source_model, scenario, ckpt)

        # Observation noise
        z += pd.Series(np.random.normal(0, noise_sd, len(FEATURES)), index=FEATURES)

        row = {
            "scenario": scenario,
            "source_model": source_model,
            "source_type": "Human" if source_model == "Human" else "AI",
            "tradition": tradition,
            "axis": axis,
            "unit_id": unit_id,
            "checkpoint": ckpt,
            "checkpoint_scaled": ckpt / 15.0
        }
        row.update(z.to_dict())
        rows.append(row)

    return rows

def simulate_dataset():
    rows = []
    for scenario in SCENARIOS:
        # Human units: 2 traditions x 3 axes x 10 = 60
        for tradition in TRADITIONS:
            for axis in AXES:
                for i in range(1, N_HUMAN_PER_CELL + 1):
                    unit_id = f"{scenario}|Human|{tradition}|{axis}|P{i:02d}"
                    rows.extend(simulate_one_unit(scenario, "Human", tradition, axis, unit_id))

        # AI units: 2 AI models x 2 traditions x 3 axes x repeated runs
        for source_model in ["Claude_Proprietary", "Open_Model"]:
            for tradition in TRADITIONS:
                for axis in AXES:
                    for r in range(1, N_AI_RUNS_PER_CELL + 1):
                        unit_id = f"{scenario}|{source_model}|{tradition}|{axis}|R{r:02d}"
                        rows.extend(simulate_one_unit(scenario, source_model, tradition, axis, unit_id))

    return pd.DataFrame(rows)

df_raw = simulate_dataset()

print("\nSynthetic raw dataset shape:", df_raw.shape)
display(df_raw.head())

df_raw.to_csv(os.path.join(TABLE_DIR, "synthetic_raw_response_features.csv"), index=False)

# ----------------------------
# 5. Standardisation within scenario
# ----------------------------

df = df_raw.copy()

for scenario in SCENARIOS:
    mask = df["scenario"] == scenario
    scaler = StandardScaler()
    df.loc[mask, [f"z_{f}" for f in FEATURES]] = scaler.fit_transform(df.loc[mask, FEATURES])

Z_FEATURES = [f"z_{f}" for f in FEATURES]

# ----------------------------
# 6. Emotional Representative Distance and ADI
# ----------------------------

BLOCK_WEIGHTS = {
    "Lexical": 1/6,
    "Syntactic": 1/6,
    "Affective_VAD": 1/6,
    "Fine_Emotion": 1/6,
    "Embedding": 1/6,
    "Rasa_Bhava": 1/6
}

def cosine_distance(a, b, eps=1e-12):
    a = np.asarray(a, dtype=float)
    b = np.asarray(b, dtype=float)
    return 1.0 - np.dot(a, b) / ((np.linalg.norm(a) + eps) * (np.linalg.norm(b) + eps))

def block_distance(row_vec, base_vec, block):
    feats = [f"z_{x}" for x in FEATURE_BLOCKS[block]]
    a = row_vec[feats].values.astype(float)
    b = base_vec[feats].values.astype(float)

    if block == "Embedding":
        return cosine_distance(a, b)
    else:
        return np.linalg.norm(a - b) / np.sqrt(len(feats))

def emotional_representative_distance(row_vec, base_vec):
    block_ds = {}
    total = 0.0
    for block, w in BLOCK_WEIGHTS.items():
        d = block_distance(row_vec, base_vec, block)
        block_ds[block] = d
        total += w * (d ** 2)
    return np.sqrt(total), block_ds

# baseline lookup for each unit
baseline_df = df[df["checkpoint"] == 0].set_index(["scenario", "unit_id"])

adi_values = []
block_values = {f"ADI_block_{block}": [] for block in FEATURE_BLOCKS.keys()}

for idx, row in df.iterrows():
    base = baseline_df.loc[(row["scenario"], row["unit_id"])]
    erd, block_ds = emotional_representative_distance(row, base)
    if row["checkpoint"] == 0:
        erd = 0.0
        block_ds = {b: 0.0 for b in FEATURE_BLOCKS.keys()}
    adi_values.append(erd)
    for block in FEATURE_BLOCKS.keys():
        block_values[f"ADI_block_{block}"].append(block_ds[block])

df["ADI"] = adi_values
for col, vals in block_values.items():
    df[col] = vals

df["log1p_ADI"] = np.log1p(df["ADI"])

# Feature deltas from baseline
for f in FEATURES:
    zf = f"z_{f}"
    delta_col = f"delta_{f}"
    df[delta_col] = np.nan

baseline_vectors = df[df["checkpoint"] == 0].set_index(["scenario", "unit_id"])[Z_FEATURES]

for idx, row in df.iterrows():
    base_vec = baseline_vectors.loc[(row["scenario"], row["unit_id"])]
    for f in FEATURES:
        df.loc[idx, f"delta_{f}"] = row[f"z_{f}"] - base_vec[f"z_{f}"]

df.to_csv(os.path.join(TABLE_DIR, "synthetic_dataset_with_ADI_and_deltas.csv"), index=False)

print("\nDataset with ADI shape:", df.shape)
display(df[["scenario", "source_model", "tradition", "axis", "unit_id", "checkpoint", "ADI"]].head(12))

# ----------------------------
# 7. Descriptive tables
# ----------------------------

design_table = (
    df.drop_duplicates(["scenario", "source_model", "tradition", "axis", "unit_id"])
      .groupby(["scenario", "source_model", "tradition", "axis"])
      .size()
      .reset_index(name="number_of_units")
)

feature_table = []
for block, feats in FEATURE_BLOCKS.items():
    for f in feats:
        feature_table.append({"block": block, "feature": f})
feature_table = pd.DataFrame(feature_table)

summary_table = (
    df.groupby(["scenario", "source_model", "tradition", "axis", "checkpoint"])
      .agg(
          mean_ADI=("ADI", "mean"),
          sd_ADI=("ADI", "std"),
          n=("ADI", "size")
      )
      .reset_index()
)

final_summary_table = (
    df[df["checkpoint"] == 15]
    .groupby(["scenario", "source_model", "tradition"])
    .agg(
        mean_final_ADI=("ADI", "mean"),
        sd_final_ADI=("ADI", "std"),
        n=("ADI", "size")
    )
    .reset_index()
    .sort_values(["scenario", "source_model", "tradition"])
)

design_table.to_csv(os.path.join(TABLE_DIR, "design_table.csv"), index=False)
feature_table.to_csv(os.path.join(TABLE_DIR, "feature_dictionary.csv"), index=False)
summary_table.to_csv(os.path.join(TABLE_DIR, "ADI_summary_by_group_checkpoint.csv"), index=False)
final_summary_table.to_csv(os.path.join(TABLE_DIR, "final_checkpoint_ADI_summary.csv"), index=False)

print("\n=== Design Table ===")
display(design_table.head(20))

print("\n=== Feature Dictionary ===")
display(feature_table)

print("\n=== Final Checkpoint ADI Summary ===")
display(final_summary_table)

# ----------------------------
# 8. Mixed-effects modelling / OLS fallback
# ----------------------------

model_rows = []

for scenario in SCENARIOS:
    dsub = df[(df["scenario"] == scenario) & (df["checkpoint"] > 0)].copy()

    formula = "log1p_ADI ~ checkpoint_scaled * C(tradition) * C(source_model) + C(axis)"

    print(f"\nFitting model for: {scenario}")

    try:
        # Random intercept by unit_id
        md = smf.mixedlm(formula, dsub, groups=dsub["unit_id"])
        mdf = md.fit(reml=False, method="lbfgs", maxiter=300, disp=False)

        params = mdf.params
        pvals = mdf.pvalues

        model_txt = mdf.summary().as_text()
        model_type = "MixedLM_random_intercept"

        with open(os.path.join(MODEL_DIR, f"mixed_model_{scenario}.txt"), "w") as f:
            f.write(model_txt)

    except Exception as e:
        print("MixedLM failed; using robust OLS fallback.")
        print("Reason:", str(e)[:300])

        ols = smf.ols(formula, dsub).fit(cov_type="HC3")
        params = ols.params
        pvals = ols.pvalues

        model_txt = ols.summary().as_text()
        model_type = "OLS_HC3_fallback"

        with open(os.path.join(MODEL_DIR, f"ols_fallback_{scenario}.txt"), "w") as f:
            f.write(model_txt)

    for name in params.index:
        model_rows.append({
            "scenario": scenario,
            "model_type": model_type,
            "term": name,
            "estimate": params[name],
            "p_value": pvals.get(name, np.nan)
        })

model_table = pd.DataFrame(model_rows)
model_table.to_csv(os.path.join(TABLE_DIR, "model_coefficients_mixed_or_ols.csv"), index=False)

print("\n=== Model Coefficients: Key Terms Containing checkpoint_scaled ===")
key_model_table = model_table[model_table["term"].str.contains("checkpoint_scaled", regex=False)]
display(key_model_table)

# ----------------------------
# 9. PERMANOVA-style distance-based group test
# ----------------------------

def pairwise_erd_matrix(data):
    """
    Pairwise emotional-representative distance using standardized vectors.
    For speed, this uses Euclidean distance on all standardized features.
    This is acceptable for demonstration; the ADI above used block-aware ERD.
    """
    X = data[Z_FEATURES].values.astype(float)
    return squareform(pdist(X, metric="euclidean"))

def permanova_oneway(distance_matrix, groups, n_perm=499, seed=2026):
    """
    Simple one-way PERMANOVA-style test.
    Tests whether multivariate distance profiles differ across groups.
    """
    rng = np.random.default_rng(seed)
    groups = np.asarray(groups)
    unique_groups = np.unique(groups)
    n = len(groups)

    D2 = distance_matrix ** 2

    # Total sum of squares from distances
    ss_total = D2.sum() / n

    def within_ss(g):
        ss_within = 0.0
        for ug in unique_groups:
            idx = np.where(g == ug)[0]
            if len(idx) > 1:
                ss_within += D2[np.ix_(idx, idx)].sum() / len(idx)
        return ss_within

    ss_within_obs = within_ss(groups)
    ss_between_obs = ss_total - ss_within_obs

    df_between = len(unique_groups) - 1
    df_within = n - len(unique_groups)

    ms_between = ss_between_obs / max(df_between, 1)
    ms_within = ss_within_obs / max(df_within, 1)
    F_obs = ms_between / (ms_within + 1e-12)
    R2_obs = ss_between_obs / (ss_total + 1e-12)

    perm_F = []
    for _ in range(n_perm):
        perm_g = rng.permutation(groups)
        ss_within_perm = within_ss(perm_g)
        ss_between_perm = ss_total - ss_within_perm
        msb = ss_between_perm / max(df_between, 1)
        msw = ss_within_perm / max(df_within, 1)
        perm_F.append(msb / (msw + 1e-12))

    perm_F = np.array(perm_F)
    p_value = (1 + np.sum(perm_F >= F_obs)) / (1 + n_perm)

    return {
        "F": F_obs,
        "R2": R2_obs,
        "p_value": p_value,
        "n_perm": n_perm,
        "n": n,
        "groups": len(unique_groups)
    }

permanova_rows = []

for scenario in SCENARIOS:
    dsub = df[(df["scenario"] == scenario) & (df["checkpoint"] == 15)].copy()
    dsub["group_source_tradition"] = dsub["source_model"] + "_" + dsub["tradition"]

    D = pairwise_erd_matrix(dsub)
    res = permanova_oneway(D, dsub["group_source_tradition"].values, n_perm=499, seed=2026)

    row = {"scenario": scenario, "grouping": "source_model_by_tradition_at_checkpoint_15"}
    row.update(res)
    permanova_rows.append(row)

permanova_table = pd.DataFrame(permanova_rows)
permanova_table.to_csv(os.path.join(TABLE_DIR, "PERMANOVA_style_results.csv"), index=False)

print("\n=== PERMANOVA-style Multivariate Group Separation Results ===")
display(permanova_table)

# ----------------------------
# 10. FDR correction for feature-level contrasts
# ----------------------------

def fdr_feature_tests(data, group_col, group_a, group_b, label):
    """
    Welch t-tests on feature deltas at final checkpoint, followed by BH-FDR.
    """
    rows = []
    d = data[data["checkpoint"] == 15].copy()

    A = d[d[group_col] == group_a]
    B = d[d[group_col] == group_b]

    for f in FEATURES:
        col = f"delta_{f}"
        if len(A) > 2 and len(B) > 2:
            stat, p = ttest_ind(A[col].values, B[col].values, equal_var=False, nan_policy="omit")
            mean_diff = A[col].mean() - B[col].mean()
        else:
            stat, p, mean_diff = np.nan, np.nan, np.nan

        rows.append({
            "contrast": label,
            "feature": f,
            "block": FEATURE_TO_BLOCK[f],
            "mean_diff_groupA_minus_groupB": mean_diff,
            "t_stat": stat,
            "raw_p_value": p
        })

    out = pd.DataFrame(rows)
    valid = out["raw_p_value"].notna()
    reject, p_adj, _, _ = multipletests(out.loc[valid, "raw_p_value"], method="fdr_bh", alpha=0.05)
    out.loc[valid, "BH_FDR_p_value"] = p_adj
    out.loc[valid, "BH_FDR_significant_0_05"] = reject
    return out

fdr_all = []

for scenario in SCENARIOS:
    dsc = df[df["scenario"] == scenario].copy()

    # Contrast 1: Human Tagore vs Human Wordsworth
    d_human = dsc[dsc["source_model"] == "Human"].copy()
    fdr_all.append(
        fdr_feature_tests(
            d_human,
            group_col="tradition",
            group_a="Tagore",
            group_b="Wordsworth",
            label=f"{scenario}: Human_Tagore_minus_Human_Wordsworth"
        )
    )

    # Contrast 2: Tagore Human vs Tagore Claude
    d_tagore = dsc[dsc["tradition"] == "Tagore"].copy()
    fdr_all.append(
        fdr_feature_tests(
            d_tagore,
            group_col="source_model",
            group_a="Human",
            group_b="Claude_Proprietary",
            label=f"{scenario}: Tagore_Human_minus_Tagore_Claude"
        )
    )

    # Contrast 3: Tagore Claude vs Tagore Open Model
    d_tagore_ai = dsc[(dsc["tradition"] == "Tagore") & (dsc["source_model"].isin(["Claude_Proprietary", "Open_Model"]))].copy()
    fdr_all.append(
        fdr_feature_tests(
            d_tagore_ai,
            group_col="source_model",
            group_a="Claude_Proprietary",
            group_b="Open_Model",
            label=f"{scenario}: Tagore_Claude_minus_Tagore_OpenModel"
        )
    )

fdr_table = pd.concat(fdr_all, ignore_index=True)
fdr_table.to_csv(os.path.join(TABLE_DIR, "feature_level_FDR_tests.csv"), index=False)

sig_counts = (
    fdr_table.groupby(["contrast", "block"])
    .agg(significant_features=("BH_FDR_significant_0_05", "sum"))
    .reset_index()
)

sig_counts.to_csv(os.path.join(TABLE_DIR, "FDR_significant_feature_counts_by_block.csv"), index=False)

print("\n=== FDR Feature-Level Tests: First Rows ===")
display(fdr_table.head(20))

print("\n=== Significant Feature Counts by Block ===")
display(sig_counts.head(30))

# ----------------------------
# 11. Plotting helpers
# ----------------------------

plt.rcParams.update({
    "figure.figsize": (10, 6),
    "axes.titlesize": 14,
    "axes.labelsize": 12,
    "legend.fontsize": 9,
    "xtick.labelsize": 10,
    "ytick.labelsize": 10,
    "savefig.dpi": 220,
    "figure.dpi": 120
})

def save_and_show(fig, filename):
    path = os.path.join(FIG_DIR, filename)
    fig.tight_layout()
    fig.savefig(path, bbox_inches="tight")
    plt.show()
    plt.close(fig)
    return path

def plot_adi_trajectories(scenario):
    d = summary_table[summary_table["scenario"] == scenario].copy()

    fig, ax = plt.subplots()
    for (sm, tr), sub in d.groupby(["source_model", "tradition"]):
        s = sub.groupby("checkpoint")["mean_ADI"].mean().reset_index()
        ax.plot(s["checkpoint"], s["mean_ADI"], marker="o", label=f"{sm} | {tr}")

    ax.set_title(f"Affective Drift Index Trajectories\n{scenario}")
    ax.set_xlabel("Exposure checkpoint")
    ax.set_ylabel("Mean Affective Drift Index")
    ax.grid(True, alpha=0.3)
    ax.legend(loc="best", ncol=2)
    return save_and_show(fig, f"ADI_trajectories_{scenario}.png")

def plot_velocity(scenario):
    d = summary_table[summary_table["scenario"] == scenario].copy()
    g = (
        d.groupby(["source_model", "tradition", "checkpoint"])["mean_ADI"]
        .mean()
        .reset_index()
        .sort_values(["source_model", "tradition", "checkpoint"])
    )
    g["velocity"] = g.groupby(["source_model", "tradition"])["mean_ADI"].diff() / g.groupby(["source_model", "tradition"])["checkpoint"].diff()
    g = g[g["checkpoint"] > 0]

    labels = []
    values = []
    for (sm, tr), sub in g.groupby(["source_model", "tradition"]):
        labels.append(f"{sm}\n{tr}")
        values.append(sub["velocity"].mean())

    fig, ax = plt.subplots(figsize=(11, 5.8))
    ax.bar(range(len(values)), values)
    ax.set_xticks(range(len(values)))
    ax.set_xticklabels(labels, rotation=30, ha="right")
    ax.set_title(f"Average Drift Velocity\n{scenario}")
    ax.set_ylabel("Mean velocity across checkpoint intervals")
    ax.grid(True, axis="y", alpha=0.3)
    return save_and_show(fig, f"Drift_velocity_{scenario}.png")

def plot_pca_final(scenario):
    d = df[(df["scenario"] == scenario) & (df["checkpoint"].isin([0, 15]))].copy()
    X = d[Z_FEATURES].values
    pca = PCA(n_components=2, random_state=2026)
    coords = pca.fit_transform(X)
    d["PC1"] = coords[:, 0]
    d["PC2"] = coords[:, 1]

    fig, ax = plt.subplots()
    for (sm, tr, ck), sub in d.groupby(["source_model", "tradition", "checkpoint"]):
        marker = "o" if ck == 0 else "^"
        ax.scatter(sub["PC1"], sub["PC2"], s=38, alpha=0.70, marker=marker, label=f"{sm}|{tr}|k={ck}")

    ax.set_title(f"PCA Map of Baseline and Final Response Vectors\n{scenario}")
    ax.set_xlabel(f"PC1 ({pca.explained_variance_ratio_[0]*100:.1f}% variance)")
    ax.set_ylabel(f"PC2 ({pca.explained_variance_ratio_[1]*100:.1f}% variance)")
    ax.grid(True, alpha=0.3)
    ax.legend(loc="best", fontsize=7, ncol=2)
    return save_and_show(fig, f"PCA_baseline_final_{scenario}.png")

def plot_block_contributions(scenario):
    d = df[(df["scenario"] == scenario) & (df["checkpoint"] == 15)].copy()
    block_cols = [f"ADI_block_{b}" for b in FEATURE_BLOCKS.keys()]
    g = d.groupby(["source_model", "tradition"])[block_cols].mean().reset_index()
    g["label"] = g["source_model"] + "\n" + g["tradition"]

    fig, ax = plt.subplots(figsize=(11, 6))
    bottom = np.zeros(len(g))
    for col in block_cols:
        values = g[col].values
        ax.bar(range(len(g)), values, bottom=bottom, label=col.replace("ADI_block_", ""))
        bottom += values

    ax.set_xticks(range(len(g)))
    ax.set_xticklabels(g["label"], rotation=30, ha="right")
    ax.set_title(f"Final Checkpoint Block-wise Drift Contributions\n{scenario}")
    ax.set_ylabel("Mean block distance")
    ax.grid(True, axis="y", alpha=0.3)
    ax.legend(loc="best", ncol=2)
    return save_and_show(fig, f"Block_contributions_{scenario}.png")

def plot_group_distance_heatmap(scenario):
    d = df[(df["scenario"] == scenario) & (df["checkpoint"] == 15)].copy()
    d["group"] = d["source_model"] + "_" + d["tradition"]

    centroids = d.groupby("group")[Z_FEATURES].mean()
    labels = centroids.index.tolist()
    D = squareform(pdist(centroids.values, metric="euclidean"))

    fig, ax = plt.subplots(figsize=(8, 7))
    im = ax.imshow(D)
    ax.set_xticks(range(len(labels)))
    ax.set_xticklabels(labels, rotation=45, ha="right")
    ax.set_yticks(range(len(labels)))
    ax.set_yticklabels(labels)
    ax.set_title(f"Distance Heatmap among Group Centroids\n{scenario}")
    cbar = plt.colorbar(im, ax=ax)
    cbar.set_label("Euclidean distance in standardized feature space")

    for i in range(D.shape[0]):
        for j in range(D.shape[1]):
            ax.text(j, i, f"{D[i,j]:.1f}", ha="center", va="center", fontsize=8)

    return save_and_show(fig, f"Group_centroid_distance_heatmap_{scenario}.png")

def plot_fdr_counts():
    d = sig_counts.copy()
    d = d[d["significant_features"] > 0].copy()

    if d.empty:
        print("No significant FDR features found in this synthetic run.")
        return None

    # Plot only top 25 rows for readability
    d["label"] = d["contrast"].str.replace("S0_", "S0\n", regex=False)
    d = d.sort_values("significant_features", ascending=False).head(25)

    fig, ax = plt.subplots(figsize=(12, 7))
    ax.barh(range(len(d)), d["significant_features"])
    ax.set_yticks(range(len(d)))
    ax.set_yticklabels(d["contrast"].str[:55] + " | " + d["block"])
    ax.invert_yaxis()
    ax.set_title("FDR-Significant Feature Counts by Contrast and Feature Block")
    ax.set_xlabel("Number of significant features")
    ax.grid(True, axis="x", alpha=0.3)
    return save_and_show(fig, "FDR_significant_feature_counts.png")

# Generate plots
figure_paths = []
for scenario in SCENARIOS:
    print(f"\nGenerating plots for {scenario}")
    figure_paths.append(plot_adi_trajectories(scenario))
    figure_paths.append(plot_velocity(scenario))
    figure_paths.append(plot_pca_final(scenario))
    figure_paths.append(plot_block_contributions(scenario))
    figure_paths.append(plot_group_distance_heatmap(scenario))

figure_paths.append(plot_fdr_counts())

# ----------------------------
# 12. Scenario interpretation table
# ----------------------------

scenario_interpretation = pd.DataFrame([
    {
        "scenario": "S0_Control_Minimal_Drift",
        "meaning": "Almost no systematic exposure effect. ADI should remain small. Useful as a negative-control case."
    },
    {
        "scenario": "S1_Human_Tagore_Cultural_Drift_AI_Flattening",
        "meaning": "Humans show stronger Tagore-linked cultural-affective drift, while AI responses show weaker Rasa/Bhava movement. Demonstrates possible affective flattening."
    },
    {
        "scenario": "S2_Claude_Snapback_Open_Sustained_Drift",
        "meaning": "Proprietary model drifts early but partially returns/snapbacks later; open/API model keeps drifting. Demonstrates model-specific contextual stability."
    },
    {
        "scenario": "S3_Open_Model_Unstable_Semantic_Drift",
        "meaning": "Open/API model has larger random semantic movement and variability. Demonstrates instability and repeated-run variability."
    }
])

scenario_interpretation.to_csv(os.path.join(TABLE_DIR, "scenario_interpretation.csv"), index=False)

print("\n=== Scenario Interpretation ===")
display(scenario_interpretation)

# ----------------------------
# 13. Compact reporting tables
# ----------------------------

# A compact "main findings" table from synthetic data
main_findings = []

for scenario in SCENARIOS:
    fs = final_summary_table[final_summary_table["scenario"] == scenario].copy()

    row = {"scenario": scenario}

    for sm in SOURCE_MODELS:
        for tr in TRADITIONS:
            val = fs[(fs["source_model"] == sm) & (fs["tradition"] == tr)]["mean_final_ADI"]
            row[f"{sm}_{tr}_mean_final_ADI"] = float(val.iloc[0]) if len(val) else np.nan

    pval = permanova_table[permanova_table["scenario"] == scenario]["p_value"].iloc[0]
    r2 = permanova_table[permanova_table["scenario"] == scenario]["R2"].iloc[0]
    row["PERMANOVA_style_p_value"] = pval
    row["PERMANOVA_style_R2"] = r2

    main_findings.append(row)

main_findings = pd.DataFrame(main_findings)
main_findings.to_csv(os.path.join(TABLE_DIR, "main_synthetic_findings_summary.csv"), index=False)

print("\n=== Main Synthetic Findings Summary ===")
display(main_findings)

# ----------------------------
# 14. README
# ----------------------------

readme = f"""
Synthetic Affective AI Methodology Demonstration
Generated: {datetime.datetime.now().isoformat()}

This folder contains a complete synthetic demonstration of the statistical/ML layer proposed for:
"Affective Colonialism and Emotional AI: Human--Algorithm Affective Drift through Tagore and Wordsworth."

The code creates artificial response-vector data for:
- Traditions: Tagore, Wordsworth
- Axes: Transcendental, Relational, Environmental
- Sources: Human, Claude-like proprietary model, open/API model
- Checkpoints: 0, 3, 6, 9, 12, 15

Important files:
1. tables/synthetic_raw_response_features.csv
   Raw synthetic feature vectors.

2. tables/synthetic_dataset_with_ADI_and_deltas.csv
   Dataset with standardized features, deltas from baseline, and Affective Drift Index.

3. tables/design_table.csv
   Number of units by scenario/source/tradition/axis.

4. tables/ADI_summary_by_group_checkpoint.csv
   Mean and SD of ADI by group and checkpoint.

5. tables/final_checkpoint_ADI_summary.csv
   Final checkpoint ADI summaries.

6. tables/model_coefficients_mixed_or_ols.csv
   Mixed-effects model coefficients or robust OLS fallback coefficients.

7. tables/PERMANOVA_style_results.csv
   Distance-based multivariate group separation tests.

8. tables/feature_level_FDR_tests.csv
   Feature-level Welch tests with Benjamini-Hochberg FDR correction.

9. figures/
   All saved plots.

Important methodological concepts demonstrated:
- Multidimensional affective-response vector
- Lexical, syntactic, affective, embedding, and Rasa/Bhava blocks
- Emotional Representative Distance
- Affective Drift Index
- Drift velocity
- Human--AI and Tagore--Wordsworth trajectory comparison
- Mixed-effects modelling for repeated responses
- PERMANOVA-style multivariate testing
- FDR correction

Important caution:
This is synthetic demonstration data only. It is intended to explain and test the proposed pipeline.
It should not be interpreted as a real result about Tagore, Wordsworth, humans, Claude, DeepSeek, Llama, or any actual AI system.
"""

with open(os.path.join(BASE_DIR, "README.txt"), "w") as f:
    f.write(readme)

# ----------------------------
# 15. Zip all outputs
# ----------------------------

zip_path = f"{BASE_DIR}.zip"

with zipfile.ZipFile(zip_path, "w", zipfile.ZIP_DEFLATED) as zf:
    for root, dirs, files in os.walk(BASE_DIR):
        for file in files:
            full_path = os.path.join(root, file)
            rel_path = os.path.relpath(full_path, BASE_DIR)
            zf.write(full_path, arcname=rel_path)

print("\nZIP created:", zip_path)

try:
    display(FileLink(zip_path))
except Exception:
    pass

# Optional automatic Colab download
try:
    from google.colab import files
    print("\nStarting browser download...")
    files.download(zip_path)
except Exception:
    print("\nNot running in Colab, or automatic download unavailable.")
    print("You can manually download the ZIP from:", zip_path)

# ----------------------------
# 16. Final console summary for copying back
# ----------------------------

print("\n" + "="*80)
print("COPYABLE SUMMARY")
print("="*80)
print("Synthetic dataset rows:", df.shape[0])
print("Synthetic feature count:", len(FEATURES))
print("Scenarios:", ", ".join(SCENARIOS))
print("\nMain findings summary:")
print(main_findings.to_string(index=False))
print("\nPERMANOVA-style results:")
print(permanova_table.to_string(index=False))
print("\nKey mixed/OLS model terms involving exposure checkpoint:")
print(key_model_table.to_string(index=False))
print("\nOutput ZIP:", zip_path)
print("="*80)
```

```python
# ============================================================
# Export the EXACT synthetic data used in the previous analysis
# into readable text files + downloadable ZIP
# ============================================================

import os, glob, zipfile, datetime
import pandas as pd
import numpy as np

try:
    from IPython.display import display, FileLink
except Exception:
    display = print

# ------------------------------------------------------------
# 1. Locate the previous output folder
# ------------------------------------------------------------

# If BASE_DIR exists from the previous simulation, use it.
# Otherwise, find the latest folder created by the previous code.
try:
    BASE_DIR
    print("Using existing BASE_DIR:", BASE_DIR)
except NameError:
    candidates = sorted(
        glob.glob("/content/affective_ai_synthetic_demo_*"),
        key=os.path.getmtime
    )
    candidates = [c for c in candidates if os.path.isdir(c)]
    if not candidates:
        raise FileNotFoundError(
            "No previous simulation folder found. Please run the simulation code first."
        )
    BASE_DIR = candidates[-1]
    print("Found latest simulation folder:", BASE_DIR)

EXPORT_DIR = os.path.join(BASE_DIR, "exact_synthetic_data_text_export")
os.makedirs(EXPORT_DIR, exist_ok=True)

TABLE_DIR = os.path.join(BASE_DIR, "tables")

# ------------------------------------------------------------
# 2. Load exact data from memory if available; otherwise from CSV
# ------------------------------------------------------------

if "df_raw" in globals():
    raw_data = df_raw.copy()
    print("Loaded raw synthetic data from memory: df_raw")
else:
    raw_path = os.path.join(TABLE_DIR, "synthetic_raw_response_features.csv")
    raw_data = pd.read_csv(raw_path)
    print("Loaded raw synthetic data from CSV:", raw_path)

if "df" in globals():
    processed_data = df.copy()
    print("Loaded processed synthetic data from memory: df")
else:
    processed_path = os.path.join(TABLE_DIR, "synthetic_dataset_with_ADI_and_deltas.csv")
    processed_data = pd.read_csv(processed_path)
    print("Loaded processed synthetic data from CSV:", processed_path)

print("\nRaw data shape:", raw_data.shape)
print("Processed data shape:", processed_data.shape)

# ------------------------------------------------------------
# 3. Identify feature columns and useful metadata
# ------------------------------------------------------------

meta_cols = [
    "scenario", "source_model", "source_type", "tradition",
    "axis", "unit_id", "checkpoint", "checkpoint_scaled"
]

raw_feature_cols = [c for c in raw_data.columns if c not in meta_cols]

adi_cols = [
    "scenario", "source_model", "source_type", "tradition",
    "axis", "unit_id", "checkpoint", "checkpoint_scaled", "ADI"
]

adi_cols = [c for c in adi_cols if c in processed_data.columns]

block_cols = [c for c in processed_data.columns if c.startswith("ADI_block_")]
delta_cols = [c for c in processed_data.columns if c.startswith("delta_")]
z_cols = [c for c in processed_data.columns if c.startswith("z_")]

# ------------------------------------------------------------
# 4. Save full exact data as tab-separated readable TXT
# ------------------------------------------------------------

raw_txt_path = os.path.join(EXPORT_DIR, "FULL_exact_raw_synthetic_data_3744x41.txt")
processed_txt_path = os.path.join(EXPORT_DIR, "FULL_exact_processed_synthetic_data_3744x115.txt")

raw_data.to_csv(raw_txt_path, sep="\t", index=False)
processed_data.to_csv(processed_txt_path, sep="\t", index=False)

print("\nSaved full raw data TXT:", raw_txt_path)
print("Saved full processed data TXT:", processed_txt_path)

# Also save CSV copies for safety
raw_csv_path = os.path.join(EXPORT_DIR, "FULL_exact_raw_synthetic_data.csv")
processed_csv_path = os.path.join(EXPORT_DIR, "FULL_exact_processed_synthetic_data.csv")

raw_data.to_csv(raw_csv_path, index=False)
processed_data.to_csv(processed_csv_path, index=False)

# ------------------------------------------------------------
# 5. Save smaller human-readable preview files
# ------------------------------------------------------------

def save_text(path, text):
    with open(path, "w", encoding="utf-8") as f:
        f.write(text)

# General explanation
summary_text = f"""
EXACT SYNTHETIC DATA USED IN THE SIMULATION ANALYSIS

This folder exports the same synthetic data that were used in the previous simulation analysis.

Important note:
These are not real human or AI text responses. They are synthetic numerical response-feature vectors generated only to demonstrate the proposed methodology.

Raw synthetic data:
- Rows: {raw_data.shape[0]}
- Columns: {raw_data.shape[1]}
- File: FULL_exact_raw_synthetic_data_3744x41.txt

Processed synthetic data:
- Rows: {processed_data.shape[0]}
- Columns: {processed_data.shape[1]}
- File: FULL_exact_processed_synthetic_data_3744x115.txt

The processed file contains:
- raw metadata,
- standardized features z_*,
- baseline difference features delta_*,
- block-wise Affective Drift Index components,
- final Affective Drift Index, ADI.

Scenarios:
{chr(10).join('- ' + str(x) for x in raw_data['scenario'].unique())}

Source models:
{chr(10).join('- ' + str(x) for x in raw_data['source_model'].unique())}

Traditions:
{chr(10).join('- ' + str(x) for x in raw_data['tradition'].unique())}

Affective axes:
{chr(10).join('- ' + str(x) for x in raw_data['axis'].unique())}

Checkpoints:
{', '.join(map(str, sorted(raw_data['checkpoint'].unique())))}

Raw synthetic feature columns:
{chr(10).join('- ' + str(x) for x in raw_feature_cols)}
"""

save_text(os.path.join(EXPORT_DIR, "README_exact_synthetic_data.txt"), summary_text)

# Pretty preview: first 40 rows of raw data
with pd.option_context(
    "display.max_rows", 50,
    "display.max_columns", None,
    "display.width", 240,
    "display.max_colwidth", 80
):
    raw_preview = raw_data.head(40).to_string(index=False)
    processed_preview = processed_data[adi_cols + block_cols].head(60).to_string(index=False)

save_text(
    os.path.join(EXPORT_DIR, "PREVIEW_first_40_rows_raw_data.txt"),
    raw_preview
)

save_text(
    os.path.join(EXPORT_DIR, "PREVIEW_first_60_rows_processed_ADI_data.txt"),
    processed_preview
)

# ------------------------------------------------------------
# 6. Save one compact sample per scenario for visual inspection
# ------------------------------------------------------------

for scenario in raw_data["scenario"].unique():
    safe_name = scenario.replace("/", "_").replace(" ", "_")

    sample_raw = (
        raw_data[raw_data["scenario"] == scenario]
        .sort_values(["source_model", "tradition", "axis", "unit_id", "checkpoint"])
        .head(80)
    )

    sample_processed = (
        processed_data[processed_data["scenario"] == scenario]
        .sort_values(["source_model", "tradition", "axis", "unit_id", "checkpoint"])
        .head(80)
    )

    with pd.option_context(
        "display.max_rows", 100,
        "display.max_columns", None,
        "display.width", 260,
        "display.max_colwidth", 70
    ):
        save_text(
            os.path.join(EXPORT_DIR, f"SAMPLE_raw_data_{safe_name}.txt"),
            sample_raw.to_string(index=False)
        )

        keep_cols = adi_cols + block_cols
        keep_cols = [c for c in keep_cols if c in sample_processed.columns]
        save_text(
            os.path.join(EXPORT_DIR, f"SAMPLE_processed_ADI_data_{safe_name}.txt"),
            sample_processed[keep_cols].to_string(index=False)
        )

# ------------------------------------------------------------
# 7. Save summary tables again in readable TXT form
# ------------------------------------------------------------

summary_files = [
    "design_table.csv",
    "feature_dictionary.csv",
    "ADI_summary_by_group_checkpoint.csv",
    "final_checkpoint_ADI_summary.csv",
    "main_synthetic_findings_summary.csv",
    "PERMANOVA_style_results.csv",
    "model_coefficients_mixed_or_ols.csv",
    "FDR_significant_feature_counts_by_block.csv"
]

for fname in summary_files:
    fpath = os.path.join(TABLE_DIR, fname)
    if os.path.exists(fpath):
        tmp = pd.read_csv(fpath)
        txt_name = fname.replace(".csv", ".txt")
        with pd.option_context(
            "display.max_rows", 300,
            "display.max_columns", None,
            "display.width", 260,
            "display.max_colwidth", 100
        ):
            save_text(
                os.path.join(EXPORT_DIR, txt_name),
                tmp.to_string(index=False)
            )

# ------------------------------------------------------------
# 8. Print important previews inline
# ------------------------------------------------------------

print("\n" + "="*90)
print("RAW SYNTHETIC DATA: FIRST 10 ROWS")
print("="*90)
display(raw_data.head(10))

print("\n" + "="*90)
print("PROCESSED DATA WITH ADI: FIRST 12 ROWS")
print("="*90)
display(processed_data[adi_cols + block_cols].head(12))

print("\n" + "="*90)
print("FINAL CHECKPOINT SUMMARY, IF AVAILABLE")
print("="*90)

summary_path = os.path.join(TABLE_DIR, "main_synthetic_findings_summary.csv")
if os.path.exists(summary_path):
    main_summary = pd.read_csv(summary_path)
    display(main_summary)
else:
    print("main_synthetic_findings_summary.csv not found.")

# ------------------------------------------------------------
# 9. Create downloadable ZIP
# ------------------------------------------------------------

zip_path = os.path.join(BASE_DIR, "exact_synthetic_data_text_export.zip")

with zipfile.ZipFile(zip_path, "w", zipfile.ZIP_DEFLATED) as zf:
    for root, dirs, files in os.walk(EXPORT_DIR):
        for file in files:
            full_path = os.path.join(root, file)
            rel_path = os.path.relpath(full_path, EXPORT_DIR)
            zf.write(full_path, arcname=rel_path)

print("\nZIP created:", zip_path)

try:
    display(FileLink(zip_path))
except Exception:
    pass

try:
    from google.colab import files
    print("Starting browser download...")
    files.download(zip_path)
except Exception:
    print("Automatic Colab download unavailable. Download manually from:", zip_path)
```
